```python
### This does require " pip install pyfonts " in your environment
###
### Show all of the CURRENTLY available fonts on the system
import matplotlib.pyplot as plt
import matplotlib.font_manager
matplotlib.font_manager.findSystemFonts(fontpaths=None, fontext='ttf')
```

```python
### You can import fonts from a directory. 
### I setup a folder on "project" which has all of the Google fonts, plus Arial and Helvetica

font1 = matplotlib.font_manager.FontProperties(fname="/project/jabrzenski/Fonts/fonts/ttf/Helvetica.ttf")
font2 = matplotlib.font_manager.FontProperties(fname="/project/jabrzenski/Fonts/fonts/ttf/Arial.ttf")
font3 = matplotlib.font_manager.FontProperties(fname="/project/jabrzenski/Fonts/fonts/ttf/Crushed-Regular.ttf")
font4 = matplotlib.font_manager.FontProperties(fname="/project/jabrzenski/Fonts/fonts/ttf/CreepsterCaps-Regular.ttf")

fig, ax = plt.subplots(figsize=(6, 3))

# Add text with different fonts
ax.text(0.5, 0.7, "This is Font 1 - Helvetica",  fontproperties=font1, fontsize=20, ha="center")
ax.text(0.5, 0.4, "This is Font 2 - Arial?!?",   fontproperties=font2, fontsize=20, ha="center")
ax.text(0.5, 0.1, "This is Font 3 - Crushed?!?", fontproperties=font3, fontsize=20, ha="center")
ax.text(0.5,-0.2, "This is Font 4 - Creepster?!?", fontproperties=font4, fontsize=20, ha="center")

ax.set_xticks([])
ax.set_yticks([])
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_visible(False)
ax.spines["bottom"].set_visible(False)

plt.show()
```

```python
import numpy as np

x = np.arange(0, 10, 0.1)
y = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x, y)

ax.set_xlabel('Time', fontproperties=font1, fontsize=12)
ax.set_ylabel('Intensity', fontproperties=font2, fontsize=12)
ax.set_title('Sine Wave', fontproperties=font3, fontsize=14)

xticks = ax.get_xticks()
yticks = ax.get_yticks()

ax.set_xticks(xticks)
ax.set_yticks(yticks)

# Apply the custom font to tick labels
ax.set_xticklabels([f"{tick}" for tick in xticks], fontproperties=font4)
ax.set_yticklabels([f"{tick}" for tick in yticks], fontproperties=font3)

plt.show()
```

```python
## This shows all available fonts in the project folder.
import os

# Define the font directory
font_dir = "/project/jabrzenski/Fonts/fonts/ttf"

# Get all font files
font_files = sorted(
    [os.path.join(font_dir, f) for f in os.listdir(font_dir) if f.endswith((".ttf", ".otf", "TTF"))]
)

# Display each font
fig, ax = plt.subplots(len(font_files), 1, figsize=(8, len(font_files)*1.5))

for i, font_path in enumerate(font_files):
    font = matplotlib.font_manager.FontProperties(fname=font_path)
    ax[i].text(0.5, 0.8, "FontPartyFolder!!", fontproperties=font, fontsize=15, ha="center", va="center")
    ax[i].set_title(os.path.basename(font_path), fontsize=12)
    ax[i].axis("off")

plt.show()
```
